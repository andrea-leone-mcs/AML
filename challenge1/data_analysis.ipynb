```python
r_yes=[]
r_no=[]
g_yes=[]
g_no=[]
b_yes=[]
b_no=[]
path_train=r'data/train.csv'
df=pd.read_csv(path_train)
path=r"data/train/train"
for index, row in df.iterrows():
    
    
    image = cv2.imread(os.path.join(path, row['id']))
    
    
    # Convert the image from BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Compute the average values for each channel
    if row['has_cactus']==1:
        r_yes.append(image_rgb[:,:,0].flatten())
        g_yes.append(image_rgb[:,:,1].flatten())
        b_yes.append(image_rgb[:,:,2].flatten())
    else:
        r_no.append(image_rgb[:,:,0].flatten())
        g_no.append(image_rgb[:,:,1].flatten())
        b_no.append(image_rgb[:,:,2].flatten())
    
    
```

```python
r1=np.concatenate(r_yes).flatten()
r2=np.concatenate(r_no).flatten()
g1=np.concatenate(g_yes).flatten()
g2=np.concatenate(g_no).flatten()
b1=np.concatenate(b_yes).flatten()
b2=np.concatenate(b_no).flatten()
```

```python
# Plot the normalized histograms
from matplotlib import pyplot as plt
plt.hist(r1, bins=256, range=[0, 256], color='red', alpha=0.7, density=True, label='yes')
plt.hist(r2, bins=256, range=[0, 256], color='blue', alpha=0.7, density=True, label='no')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Normalized Frequency')
plt.title('Normalized Histogram')
plt.legend()

# Show plot
plt.show()
plt.show()
```
